5. Data Imputation in the Retail Sales Dataset
   - Task: Handle missing values in the Retail Sales dataset using advanced imputation techniques like KNN imputation and MICE.
   - Dataset: Retail Sales Dataset


```python
import pandas as pd
```

```python
df=pd.read_csv('retail_sales_dataset.csv')
df.head()
```

```python
df.isna().sum()
```

```python
import numpy as np

# Introduce missing values
df_missing = df.copy()
df_missing.loc[1, 'Age'] = np.nan
df_missing.loc[3, 'Total Amount'] = np.nan

print("Dataset with Missing Values:\n", df_missing)
```

```python
df_missing
```

```python
df_missing.isna().sum()
```

```python
from sklearn.impute import KNNImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
numerical_cols
```

```python
knn_imputer=KNNImputer(n_neighbors=2)
```

```python
df_numeric=df_missing[numerical_cols]
```

```python
df_knn=pd.DataFrame(knn_imputer.fit_transform(df_numeric),columns=numerical_cols)
```

```python
df_knn
```

```python
df_missing[numerical_cols] = df_knn
df_missing[categorical_cols] =df_missing[categorical_cols]
```

```python
df_new=df_missing
```
