```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Drop rows with missing values in 'age'
df_age = df.dropna(subset=['age'])

# Count the frequency of each class
class_counts = df['class'].value_counts()

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Histogram of ages
axs[0].hist(df_age['age'], bins=20, color='skyblue')
axs[0].set_title('Histogram of Ages')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')

# Bar chart of class frequencies
axs[1].bar(class_counts.index, class_counts.values, color='salmon')
axs[1].set_title('Frequency of Passengers in Each Class')
axs[1].set_xlabel('Class')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
