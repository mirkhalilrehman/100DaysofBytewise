```python
import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset('titanic')

df = df.dropna(subset=['age'])


sns.histplot(df['age'], kde=False, color='skyblue', label='Histogram')


sns.kdeplot(df['age'], color='red', label='KDE')

plt.title('Histogram and KDE of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
