Exercise: Create a simple line plot using Matplotlib to display the trend of a list of values over time.

```python
import matplotlib.pyplot as plt

# Sample data
values = [1, 3, 2, 5, 7, 8, 6, 4, 5, 9, 11]
time = range(len(values))

# Create a line plot
plt.plot(time, values, marker='o')
plt.title('Trend Over Time')
plt.xlabel('Time')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
